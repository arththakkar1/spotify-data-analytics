# Import pandas library
This cell imports the pandas library, which is essential for data manipulation and analysis in Python.

```python
import pandas as pd
```

# Load the dataset and preview
This cell loads the Spotify dataset from the CSV file and displays the first few rows to get an overview of the data.

```python
df=pd.read_csv("dataset.csv")
df.head()
```

# Check data types
This cell displays the data types of each column to ensure they are appropriate for analysis.

```python
df.dtypes
```

# Check dataset shape
This cell shows the number of rows and columns in the dataset, giving an idea of its size.

```python
df.shape
```

# Remove unnecessary columns
This cell removes the 'Unnamed: 0' column, which is likely an index column added during CSV export and not needed for analysis.

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
```

# Check for missing values
This cell checks each column for missing values, which is important for data cleaning and ensuring data quality.

```python
df.isnull().sum()
```

# Remove duplicate rows
This cell removes any duplicate records from the dataset to ensure each entry is unique.

```python
df.drop_duplicates(inplace=True)
```

# Standardize string columns
This cell trims whitespace and converts text to lowercase for track name, album name, and artists columns to ensure consistency.

```python
df['track_name'] = df['track_name'].str.strip().str.lower()
df['album_name'] = df['album_name'].str.strip().str.lower()
df['artists'] = df['artists'].str.strip().str.lower()
```

# Correct data types
This cell ensures that the 'explicit' column is boolean and 'duration_ms' is integer for accurate analysis.

```python
df['explicit'] = df['explicit'].astype(bool)
df['duration_ms'] = df['duration_ms'].astype(int)
```

# Fill missing genres
This cell fills missing values in the 'track_genre' column with 'unknown' to handle incomplete data.

```python
df['track_genre'].fillna('unknown', inplace=True)
```

# Split and clean artists column
This cell splits the 'artists' column into individual artists, explodes the list into separate rows, and trims whitespace for consistency.

```python
df['artists'] = df['artists'].str.split(',')
df = df.explode('artists')
df['artists'] = df['artists'].str.strip()
```

# Create duration in minutes
This cell creates a new column 'duration_min' by converting the duration from milliseconds to minutes for easier interpretation.

```python
df['duration_min'] = df['duration_ms'] / 60000
```

# Standardize genre names
This cell standardizes the 'track_genre' column by converting all genre names to lowercase and removing extra spaces.

```python
df['track_genre'] = df['track_genre'].str.lower().str.strip()
```

# Filter out invalid durations
This cell removes any tracks with a duration of zero or less, ensuring only valid tracks remain in the dataset.

```python
df = df[df['duration_ms'] > 0]
```

# Export cleaned data
This cell exports the cleaned DataFrame to a new CSV file named 'cleaned_spotify.csv' for future use or analysis.

```python
df.to_csv("cleaned_spotify.csv", index=False)
```
